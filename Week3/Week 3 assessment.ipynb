```python
'''
1) Write a Pandas program to convert a given dictionary into a Dataframe.
The first element of the input list must be the column name.
The converted dataframe should be returned.
'''
def to_dict(dict):
    import pandas as pd
    return pd.DataFrame.from_dict(dict)
```

```python
'''
2b) Write inside the makeCategory() function to only return a column called 'price_category' 
which categorizes the price_range as follows:

"Below $50": Cheap

"$51-100": Not Cheap

"$101-200": Expensive

"$201-400": Very Expensive
'''
def Category(x):
    scale = {"Below $50":"Cheap",
         "$51-100":"Not Cheap",
         "$101-200":"Expensive",
         "$201-400":"Very Expensive"
        }
    return scale[x]

def makeCategory(x):
    x["price_category"] = x["price_range"].apply(Category)
    return x.price_category
```

```python
'''
2c) Write inside the totalDislike() function to only return the total number of dislikes for each price range.
'''
def totalDislike(x):
    return x.groupby('price_range')["price_range","dislikes"].agg('sum').reset_index()
```

```python
'''
2d) Write inside the totalBookmarks() function to return the sum of bookmarks of all 'Hong Kong Style' restaurants.
'''
def totalBookmarks(x):
    a = x["food_type"] == "Hong Kong Style"
    return x[a].bookmarks.sum()
```
